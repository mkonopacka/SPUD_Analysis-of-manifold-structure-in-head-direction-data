# Isomap example with digits dataset

```python
from sklearn.datasets import load_digits
from sklearn.manifold import Isomap
import matplotlib.pyplot as plt
```

## Dataset with 8x8 images

```python
X, labels = load_digits(return_X_y=True)
X.shape
```

```python
labels[:10]
```

```python
fig, axes = plt.subplots(2, 5, figsize=(10, 5))
for ax, img, label in zip(axes.flat, X, labels):
    ax.set(xticks=[], yticks=[])
    ax.imshow(img.reshape(8, 8), cmap=plt.cm.gray_r)
    ax.set_title(label)

plt.tight_layout()
```

## Digit embeddings in 2d space

```python
embedding = Isomap(n_components=2)
X_transformed = embedding.fit_transform(X)
X_transformed.shape
```

```python
# plot of all points in X_transformed, 2d embedding of X
plt.scatter(X_transformed[:, 0], X_transformed[:, 1], c=labels, cmap=plt.cm.get_cmap('tab10', 10), alpha=0.5)
plt.colorbar(ticks=range(10))
plt.grid()
plt.title('Isomap of Digits dataset (n_components=2)')
plt.show()
```

### 3d projection

```python
# plot of all points in X_transformed, 3d embedding of X
embedding = Isomap(n_components=3)
X_transformed = embedding.fit_transform(X)
X_transformed.shape

# plot of all points in X_transformed, 3d embedding of X
from mpl_toolkits.mplot3d import Axes3D
fig = plt.figure(figsize=(10, 8))
ax = fig.add_subplot(111, projection='3d')
sc = ax.scatter(X_transformed[:, 0], X_transformed[:, 1], X_transformed[:, 2], c=labels, cmap=plt.cm.get_cmap('tab10', 10), alpha=0.5)
ax.set_title('Isomap of Digits dataset (n_components=3)')
# plt.legend(*sc.legend_elements())
plt.colorbar(sc)
plt.show()

```
